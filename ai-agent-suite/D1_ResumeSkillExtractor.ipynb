```python
import pandas as pd
```

```python
data = {
    "resume" : [
        "Experienced python developer skilled in AI, ML, and Flask.",
        "Data analyst with experience in Excel, SQL, and Power BI",
        "Frontend engineer with React, HTML, CSS and UI design experience."
    ]
}

df = pd.DataFrame(data)
df
```

```python
skills = ["python", "ai", "ml", "flask", "sql", "excel", "react", "html", "css", "power bi", "ui design"]
```

```python
def extractSkills(text):
    found = []
    textLower = text.lower()
    for skill in skills:
        if(skill in textLower):
            found.append(skill)
    return found
```

```python
df['skillsFound'] = df['resume'].apply(extractSkills)

df
```

```python
df['score'] = df['skillsFound'].apply(lambda x: len(x))
df.sort_values(by='score', ascending=False)
```
